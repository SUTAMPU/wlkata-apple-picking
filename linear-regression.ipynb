```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Calibration dataset for x-axis
camera_x = [43, 61, 84, 80]
robot_x = [197, 215, 231, 247]

# Calibration dataset for z-axis
camera_y = [93, 105, 105, 100]
robot_z = [126, 100, 132, 100]
```

```python
# Reshaping datasets
camera_x = [[x] for x in camera_x]
robot_x = [[x] for x in robot_x]

camera_y = [[x] for x in camera_y]
robot_z = [[x] for x in robot_z]
```

```python
# X-axis
model_x = LinearRegression()
model_x.fit(camera_x, robot_x)

beta_1_x = model_x.coef_
beta_0_x = model_x.intercept_

print(f"Coefficient (beta_1): {beta_1_x}")
print(f"Intercept (beta_0): {beta_0_x}")

pred_1 = model_x.predict(camera_x)

plt.scatter(camera_x, robot_x, color='blue', label='Calibrated Data')
plt.plot(camera_x, pred_1, color='black', label='Regression Line')

plt.title('Linear Regression: Camera X-Axis vs Robot X-Axis')
plt.xlabel('Camera X-Axis (x)')
plt.ylabel('Robot X-Axis (y)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Z-axis
model_z = LinearRegression()
model_z.fit(camera_y, robot_z)

beta_1_z = model_z.coef_
beta_0_z = model_z.intercept_

print(f"Coefficient (beta_1): {beta_1_z}")
print(f"Intercept (beta_0): {beta_0_z}")

pred_2 = model_z.predict(camera_y)

plt.scatter(camera_y, robot_z, color='blue', label='Calibrated Data')
plt.plot(camera_y, pred_2, color='black', label='Regression Line')

plt.title('Linear Regression: Camera Y-Axis vs Robot Z-Axis')
plt.xlabel('Camera Y-Axis (x)')
plt.ylabel('Robot Z-Axis (y)')
plt.legend()
plt.grid(True)
plt.show()
```
